```python
import pandas as pd
import quandl, math, datetime
import numpy as np
from sklearn import preprocessing, model_selection, neighbors, svm
import matplotlib.pyplot as plt
from matplotlib import style
import pickle
```

SVM is one of the most popular ml algorithms. It is complex.

The point of SVM is to find the best separation hyperplane. It is called binary classifier which means it separtes only two groups at given time, which does not mean that it only classifies into 2 groups (it basically separtes groups into one at the time - one from the rest). The best hyperplane is when the distance between that hyperlane and the associated data that is separted is the greatest. When we have Best separation hyperlane and we want to predict group for some datapoints, we can determine on which side of the hyperlane the datapoint lies on, the side determines the group. The hyperplane may also be referred to as a decision boundary.

SVM needs linear data to perfrom well just like regression.

```python
df = pd.read_csv('./data/breast-cancer-wisconsin.data', names=['id', 'clump_thickness', 'unif_cell_size','unif_cell_shape','marg_adhesion','single_epith_cell_size','bare_nuclei','bland_chrom','norm_nucleoli','mistoses','class'])
```

```python
df.replace('?', -99999, inplace=True)
```

```python
df.drop(['id'], axis=1, inplace=True)
```

```python
X = np.array(df.drop(['class'], axis=1))
y = np.array(df['class'])
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2)
```

```python
clf = svm.SVC(kernel = "linear")
```

```python
clf.fit(X_train, y_train)
```

```python
accuracy = clf.score(X_test, y_test)
```
